```python
numbers = input("Please, input several numbers separated by comma: ")
def average(numbers):
    return float(sum(numbers))/len(numbers)
print average(numbers)
```

```python
numbers = input("Please, input several numbers separated by comma: ")
argument = input("Please, input your argument: ")
def average(numbers):
    return float(sum(numbers[-(argument):]))/int(argument)
print average(numbers)
```
